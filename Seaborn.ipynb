```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
var = np.linspace(1,10,20).reshape(4,5)
var
```

```python
v = sns.heatmap(var,vmin = 0, vmax = 20, cbar = False,xticklabels = False)
v.set(xlabel = "Java", ylabel = "Python")
plt.show()
```

```python
data =  sns.load_dataset("anagrams")
x = data.drop(columns = ["attnr"],axis = 1).head(10)
```

```python
sns.heatmap(x,vmin = 0, vmax = 10, cmap = "PuOr", annot = True)
plt.show()
```

```python
sns.heatmap(x,vmin = 0, vmax = 20)
```

```python
rav = np.linspace(1,10,10).reshape(2,5)
rav
```

```python
u = {"fontsize":20,"color":"r"}
sns.heatmap(rav, annot = True, annot_kws = u, linewidth = 10, linecolor = "r")
```

```python
ar = np.array([["a0","a1","a2","a3","a4"],
               ["b0","b1","b2","b3","b4"]])
```

```python
sns.heatmap(rav, annot = ar,cmap = "PuOr",fmt ="s")
```
