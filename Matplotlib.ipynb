```python
import matplotlib.pyplot as plt

x = [13, 23, 16, 53]
y = ["Javascript", "Java", "C++", "C"]
c = ["r", "g", "y", "b"]

# Create the pie chart
plt.pie(x, labels=y, radius=1, colors=c)

# Add a circle in the center to make it a donut chart
cr = plt.Circle(xy=(0, 0), radius=0.6, facecolor="w")  # Adjust the radius here
plt.gca().add_artist(cr)

# Show the plot
plt.show()
```

Stack Plot


```python
import matplotlib.pyplot as plt
x = [2,3,4,6,7,9]
y = [1,4,5,6,3,8]
z = [4,5,6,7,8,6]
l = ["Profit","loss"]
plt.stackplot(x,y,z,labels=l,colors=['r','y'])
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [2,3,4,6,7,9]
y = [1,4,5,6,3,9]
plt.step(x,y,label="Pyhton",marker="o",mfc="g")
plt.title("Python")
plt.legend()
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

File Between Plot


```python
import matplotlib.pyplot as plt
import numpy as np
x =np.array ([43,54,56,65])
y =np.array ([43,54,34,23])
plt.plot(x,y,color="r")
plt.fill_between(x,y,color="y",where=(x>=43) & (x<=65),alpha = 0.5)
plt.title("Python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
