```python
import tensorflow as tf
tf.compat.v1.disable_eager_execution()

a = tf.compat.v1.Variable(5)
b = tf.compat.v1.Variable(6)

c = tf.add(a,b)
new = a.assign(c)

init_op = tf.compat.v1.global_variables_initializer()

with tf.compat.v1.Session() as sess:
    sess.run(init_op)
    sess.run(new)
    print(sess.run(a))
    
```

```python
sess = tf.compat.v1.Session()

sess.run(init_op)

for i in range(5):
    sess.run(new)
    print(sess.run(a))

sess.close()
```
